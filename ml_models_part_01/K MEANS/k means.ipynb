```python
# applying emphirical rule to find the outliers
mean = data['SepalWidthCm'].mean()
std = data['SepalWidthCm'].std()
lower_limit = mean - 3*std
upper_limit = mean + 3*std
print(f'lower_limit is {lower_limit}')
print(f'upper_limit is {upper_limit}')
```

```python
# finding outliers lesser than lowerlimit
data.loc[(data['SepalWidthCm']<lower_limit)]
```

```python
# finding outliers greater than the upper limit
data.loc[(data['SepalWidthCm']>upper_limit)]
```

```python
# boxplot shows you the direction of the outliers, it doesn't mean all those points are the outliers.
```

```python
# replacing outliers with mean
data.loc[data['SepalWidthCm'] > upper_limit, "SepalWidthCm"] = data['SepalWidthCm'].mean()
```

### Model implementation:

```python
# dropping the output column and other irrelevant columns
# here only input variables
X = data
X
```

```python
# Model building

# importing KMeans model
from sklearn.cluster import KMeans

# initially random k is given as 5 then using elbow method, k is checked and taken as the value.
model = KMeans(n_clusters= 5, random_state = 10)

#training the model
model.fit(X)

# applications:
# this is used in customers segmentation
```

```python
# centroids- cluster_centers_
# no of clusters is same as number of centroids
model.cluster_centers_
# we have five clusters hence five centroids (row-wise) are taken
# four iterations happened so four columns
```

```python
# labels shows the label/ name of cluster to which a particular observation/ record/ datapoint
# belongs to .
labels= model.labels_
labels
# we have five clusters so the lables will be from 0 to 4 i.e., (0, 1, 2, 3, 4)
```

```python
# Visualizing the clusters created by the model

# different colors are given for different clusters 
color_palette = np.array(['red','blue','green','k','y'])
plt.scatter(X.SepalLengthCm, X.PetalLengthCm, color =color_palette[labels])
```

```python
# We can see which observation/record belongs to which cluster!

X[['labels']]= pd.DataFrame(labels)
X
```

```python
# We see the overlapping of clusters because the no of clusters that we have chosen might be wrong.
```

### How to find the optimum k value?
#### - Elbow method

```python
for i in range(1,11):
    print(i)
```

```python
#CREATING 'x' values and 'y' values to be plotted in the graph
# sum of squares of distances (WCSS) are calculated for all the k numbers that is 1 to 10.
wcss = []
# initializing the model and data is fitted to create 'i' no of clusters
for i in range(1,11):
    elbow = KMeans(n_clusters= i, random_state = 15)
    elbow.fit(X)
    wcss.append(elbow.inertia_) # sum of squares of distance
```

```python
wcss
```

```python
# PLOTTING THE VALUES IN THE GRAPH
# from d--r, d is diamond marker, r is red line, 
# mec- marker edge color, mfc- marker face color

plt.plot(range(1, 11), wcss, 'd--r',mec= 'g', mfc ='y')
plt.title('Elbow method')
plt.xlabel('No of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

```python
# by elbow method, correct k value is found to be 3 
# hence initalising the model again with the value found.

model1 = KMeans(n_clusters = 3, random_state = 17)
model1.fit(X)
```

```python
# cluster centers / centroids
model1.cluster_centers_

# hence 3 clusters and five iterations
```

```python
# assigning data points to the respective labels of the clusters.
model1.labels_
```

```python
# seeing the observation assigned to respective cluster in a dataframe

X[['labels']] =  pd.DataFrame(model1.labels_)
X
```

```python
# Visualizing the new clusters created by the model

palette = np.array(['red','green','blue'])
plt.scatter(data.SepalLengthCm, data.PetalLengthCm, color = palette[model1.labels_])
```

### Evaluation: 

```python
from sklearn.metrics import silhouette_score
silhouette_score(X, labels = model1.labels_)
```
