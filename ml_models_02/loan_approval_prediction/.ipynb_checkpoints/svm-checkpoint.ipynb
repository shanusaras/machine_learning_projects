## Data Preprocessing: 

### Checking null values: 
- While imputing null values for categorical features, always remember that the domain research is crucial for selecting the value to be imputed.
- here, for instance, to prevent the possible losses to the insurance company, the value which may not be the mode value can also be imputed for null values.
- eg: for dependents column, '3+' is imputed instead of the '0' dependents.

```python
data.isnull().sum()
```

```python
# handling null values
# gender

data.Gender.value_counts()
```

```python
data.loc[data.Gender.isnull()==True,'Gender']= 'Male'
```

```python
# married

data.Married.value_counts()
```

```python
data.loc[data.Married.isnull()==True,'Married']= 'Yes'
```

```python
# dependents

data.Dependents.value_counts()
```

```python
data.loc[data.Dependents.isnull()==True,'Dependents']= '3+'
```

## Encoding: 

```python
data.tail()
```

```python
# always prefer giving 1 for positive category and 0 for negative category . (eg)graduate =1, non graduate=0
```

```python
# do it for categorical columns- Gender, Married, Education, Self_Employed, Property_Area, Loan_Status
# manual encoding - can be done for certain features (to assign 1,0 values manually) - 
# label encoder- ordinal data - dependents, education
# one- hot encoder- nominal data- Gender, Married, selfemployed, property area, loan status


# any one technique can be used in accordance with the nature of the feature.
```

```python
# Gender

data['Gender'] = pd.get_dummies(data['Gender'], drop_first = True) 
```

```python
# Married

data['Married']= pd.get_dummies(data['Married'], drop_first = True)
```

```python
# self employed

data['Self_Employed']= pd.get_dummies(data['Self_Employed'], drop_first = True)
```

```python
# Loan_Status

data['Loan_Status'] =pd.get_dummies(data['Loan_Status'],drop_first = True)
```

```python
# property area

# here three unique values are present(rural, urban, semiurban)
data['Property_Area'].value_counts()

# hence three columns will be created under one hot encoder. After dropping the first column, 
# remaining two columns will be added to the original data.
```

```python
data1= pd.get_dummies(data['Property_Area'], prefix='Property_Area', drop_first = True)
```

```python
data = pd.concat([data, data1], axis= 1)
```

```python
# old general column property area is dropped
data.drop('Property_Area', axis= 1, inplace = True)
```

```python
data
```

```python
# dependents

data2 = pd.get_dummies(data['Dependents'], prefix= 'Dependents', drop_first = True)
```

```python
data= pd.concat([data, data2], axis=1)
```

```python
data.drop('Dependents', axis =1, inplace = True)
```

## Balancing the dataset:

### splitting the data 

```python
# x, y split

X= data.drop(['Loan_Status'], axis=1)
y = data['Loan_Status']
```

## Model implementation: 

### Always create the base model first with default parameters and check the accuracy.
- eg., if accuracy is 95%, there is no need for hyperparameter tuning
- if accuracy is very less, tune the hyparameters, get the optimal values, create new model with optimal
values and check the accuracy.

```python
# Support Vector Classifier model

from sklearn.svm import SVC # If support vector model is applied for regression problem, import SVR.

svm= SVC() # creating base model with default parameters (later hyper parameter tuning is done)
# press shift + tab to see the default parameters.

svm.fit(x_train_sm, y_train_sm)# training the balanced data
```

```python
# prediction

y_pred = svm.predict(x_test)
```

## Model evaluation: 

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix, f1_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
# classification report measures the quality of predictions. True positives, False positives,True 
# negatives, False negatives are used to predict the metrics of classification report.

print(classification_report(y_test, y_pred))


# precision- how many TP out of positive predictions- here 0, more errors and the model is OVERFITTED.
# recall- how many TP out of actual positives.
```

```python
# all the values in the first row is 0 which indicates that the model is overfitted.
```

```python
confusion_matrix(y_test, y_pred)
```

```python
pd.crosstab(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
# cross validation technique- this is the maximum score that a model can produce and hence this is useful
# to know the potential of the model.
# give the original data (X, y) in cross_val_score () function

# cross validation technique is more reliable for evaluation than other methods like f1 score etc.

from sklearn.model_selection import cross_val_score

scores = cross_val_score(svm, X, y, scoring='f1', cv= 3)

# splitted data is directly given into cross val score , there is no need to separately fit the data as
# done in train test split type(i.e., hold out cross validation technique)

# 'svm' is the model created using SVC earlier.
# 'f1' is taken as the evaluation metric.
# 'cv' value is given as 3 hence 3 folds will be created and scores are calculated accordingly.

print('scores:', scores)
print('cross validation score:', scores.mean())
# average scores of all iterations gives the final cross val score.

print('std:', scores.std())

# std <0.05 is good, so that there will not be much variance in all the accuracy scores from the 
# average score (i.e., all the scores are somewhat closer to the average score, for every iteration the
# model is performing somewhere closer to the average score only.)
```

### Hyperparameter tuning: 

#### To improve the model's performance :
- can do better feature engineering
- can choose better feature selection technique.
- hyperparameters can be tuned.

Every model has some hyper parameters which control the training of the model and those are what differentiating each algorithm. For SVC, 'c' and 'gamma' are the important parameters that can be tuned
to improve the model's performance.

#### Hyperparameter techniques:


- Auto hyperparameter tuning:
- 1. GridSearchCV
- 2. RandomisedSearchCV

```python
# Grid SearchCV tries all the combinations of the values passed in the dictionary and
evaluates the model for each combination using cross validation technique.

# Hence we get the accuracy for every combination of hyperparameters and
we can choose the score that gives the best performance.
```

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
# can refer the official documentation to find what are all the possible c and gamma values that could
# be given.

# verbose- a parameter which signifies how detailed the output should be
param_grid = {'C': [0.1, 5, 10, 50, 60, 70],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'random_state': (list(range(1,20)))}
model = SVC()
grid = GridSearchCV(model, param_grid, verbose =2, scoring ='f1', cv = 5)

# fitting the model for grid search
grid.fit(x_train_sm, y_train_sm)
```

```python
# printing the best parameter after tuning
print(grid.best_params_)

# printing how our model looks after hyperparameter tuning
print(grid.best_estimator_)
```

```python
# Creating a new model with optimal hyperparameter values
```

```python
model1 = SVC(C=50, gamma=1, random_state=1)
model1.fit(x_train_sm, y_train_sm)
```

```python
y_prediction = model1.predict(x_test)
```

```python
accuracy_score = (y_test, y_prediction)
accuracy_score
```

```python
print(classification_report(y_test, y_prediction))
```

```python
confusion_matrix(y_test, y_prediction)
```

```python
pd.crosstab(y_test, y_prediction)
```

```python
f1_score(y_test, y_prediction)
```

```python
# cross validation score

scores_after = cross_val_score(model1, X, y, scoring='f1', cv= 3)
print(scores_after)
print('cross validation score is', scores_after.mean())
print('std is', scores.std())
```
